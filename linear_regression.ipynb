```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('lr.csv')
df.head()
```

```python
x = np.array(df[['x']])
y = np.array(df[['y']])
x,y
```

```python
plt.scatter(x,y)
plt.xlabel('Values of x')
plt.ylabel('Values of y')
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print(x_mean,y_mean)
```

```python
num = 0
den = 0
for i in range(len(x)):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    den+=(x[i]-x_mean)**2

b1 = num/den
b0 = y_mean-(b1*x_mean)
print('Intercept:',float(b0))
print('Coefficient:',float(b1))
```

```python
y_pred = b0+b1*x
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
```
